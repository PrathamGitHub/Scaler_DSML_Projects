```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

# Import Required Libraries

In this section, we will import the necessary libraries for data manipulation and analysis, including pandas and numpy.

# Load and Inspect the Dataset

In this section, we will load the dataset containing order and shipping details into a pandas DataFrame and inspect its structure using functions like `head()`, `info()`, and `describe()`.

```python
# Load and Inspect the Dataset
# Load the dataset
file_path = 'Brand_XYZ.txt'
data = pd.read_csv(file_path, delimiter=',', encoding='latin1')

# Inspect the dataset
print(data.head())
print(data.info())
print(data.describe())
```

# Filter High-Priority Orders

In this section, we will filter the dataset to include only high-priority orders based on the `Order Priority` column.

```python
# Filter High-Priority Orders
# Filter the dataset for high-priority orders
high_priority_orders = data[data['Order Priority'] == 'Critical']

# Display the first few rows of high-priority orders
print(high_priority_orders.head())
```

# Calculate Shipping Time

In this section, we will calculate the difference in days between the `Order Date` and the `Ship Date` for high-priority orders and add it as a new column in the DataFrame.

```python
# Calculate Shipping Time
# Convert Order Date and Ship Date to datetime format
high_priority_orders['Order Date'] = pd.to_datetime(high_priority_orders['Order Date'], format='%d-%m-%Y')
high_priority_orders['Ship Date'] = pd.to_datetime(high_priority_orders['Ship Date'], format='%d-%m-%Y')

# Calculate the shipping time in days
high_priority_orders['Shipping Time (Days)'] = (high_priority_orders['Ship Date'] - high_priority_orders['Order Date']).dt.days

# Display the updated DataFrame
print(high_priority_orders[['Order Date', 'Ship Date', 'Shipping Time (Days)']].head())
```

# Compute Average Shipping Time

In this section, we will calculate the average shipping time for high-priority orders using pandas' `mean()` function and display the result.

```python
# Compute Average Shipping Time
# Calculate the average shipping time for high-priority orders
average_shipping_time = high_priority_orders['Shipping Time (Days)'].mean()

# Display the result
print(f'The average shipping time for high-priority orders is {average_shipping_time:.2f} days.')
```

# Determine Customers with Exactly 6 Purchases

In this section, we will determine how many customers have made exactly 6 purchases (orders) using the dataset.

```python
# Determine Customers with Exactly 6 Purchases
# Count the number of orders per customer
customer_order_counts = data['Customer ID'].value_counts()

# Filter customers with exactly 6 purchases
customers_with_6_purchases = customer_order_counts[customer_order_counts == 6]

# Get the number of customers with exactly 6 purchases
num_customers_with_6_purchases = len(customers_with_6_purchases)

# Display the result
print(f'The number of customers with exactly 6 purchases is {num_customers_with_6_purchases}.')
```

# Percentage Contribution of North Region to LATAM Market Sales

In this section, we will calculate the percentage contribution of the North region to the total sales in the LATAM market.

```python
# Percentage Contribution of North Region to LATAM Market Sales
# Filter the dataset for LATAM market
latam_sales = data[data['Market'] == 'LATAM']

# Calculate total sales in LATAM market
total_latam_sales = latam_sales['Sales'].sum()

# Filter the dataset for North region in LATAM market
north_latam_sales = latam_sales[latam_sales['Region'] == 'North']['Sales'].sum()

# Calculate the percentage contribution of North region to LATAM sales
north_latam_percentage = (north_latam_sales / total_latam_sales) * 100

# Display the result
print(f'The North region contributes {north_latam_percentage:.2f}% to the total sales in the LATAM market.')
```

# Average Sales for the Month and Year with the Lowest Average Profit

In this section, we will determine the month and year with the lowest average profit and calculate the average sales for that period.

```python
# Average Sales for the Month and Year with the Lowest Average Profit
# Convert Order Date to datetime format if not already done
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y')

# Extract month and year from Order Date
data['YearMonth'] = data['Order Date'].dt.to_period('M')

# Group by YearMonth and calculate average profit and sales
grouped_data = data.groupby('YearMonth').agg({'Profit': 'mean', 'Sales': 'mean'})

# Find the month and year with the lowest average profit
lowest_profit_period = grouped_data['Profit'].idxmin()
lowest_profit_sales = grouped_data.loc[lowest_profit_period, 'Sales']

# Display the result
print(f'The month and year with the lowest average profit is {lowest_profit_period}.')
print(f'The average sales for this period is {lowest_profit_sales:.2f}.')
```

# Total Units Sold for Category 'Alcohol'

In this section, we will calculate the total number of units sold for the category 'Alcohol' from the dataset `State_Category_dataset.xlsx` for the given time period.

```python
# Total Units Sold for Category 'Alcohol'
# Load the dataset
file_path = 'State_Category_dataset.xlsx'
data_alcohol = pd.read_excel(file_path)

# Inspect the dataset to verify column names
print("Column names in the dataset:")
print(data_alcohol.columns)

# Filter the dataset for category 'Alcohol'
alcohol_data = data_alcohol[data_alcohol['Category'] == 'Alcohol']

# Calculate the total units sold for 'Alcohol'
total_units_sold = alcohol_data['Unit sales'].sum()

# Display the result
print(f'The total number of units sold for the category \"Alcohol\" is {total_units_sold}.')
```

# Date of Highest Sales for the Category with the Highest Percentage Contribution

In this section, we will determine the category with the highest percentage contribution to total sales (Dollars) and find the date when the highest sales occurred for that category.

```python
# Date of Highest Sales for the Category with the Highest Percentage Contribution
# Calculate total sales for each category
category_sales = data_alcohol.groupby('Category')['Dollars'].sum()

# Find the category with the highest percentage contribution to total sales
highest_contribution_category = category_sales.idxmax()

# Filter the dataset for the highest contribution category
highest_category_data = data_alcohol[data_alcohol['Category'] == highest_contribution_category]

# Find the date with the highest sales for the category
highest_sales_date = highest_category_data.loc[highest_category_data['Dollars'].idxmax(), 'Date']

# Display the result
print(f'The category with the highest percentage contribution to total sales is \"{highest_contribution_category}\".')
print(f'The date with the highest sales for this category is {highest_sales_date}.')
```

# Category with the Highest Sales for Two Consecutive Years

In this section, we will use the `Dollars` and `Dollars last year` columns to determine which category had the highest sales for two consecutive years.

```python
# Category with the Highest Sales for Two Consecutive Years
# Group by category and calculate total sales for each year
category_sales = data_alcohol.groupby('Category').agg({'Dollars': 'sum', 'Dollars last year': 'sum'})

# Filter categories with the highest sales for both years
highest_sales_category = category_sales[(category_sales['Dollars'] == category_sales['Dollars'].max()) & \
                                        (category_sales['Dollars last year'] == category_sales['Dollars last year'].max())]

# Display the result
if not highest_sales_category.empty:
    print("The category with the highest sales for two consecutive years is:")
    print(highest_sales_category)
else:
    print("No category had the highest sales for two consecutive years.")
```

# Category with Higher Unit Sales Last Year Compared to Current Year

In this section, we will identify the category that had higher unit sales last year compared to the current year using the `Unit sales` and `Unit sales last year` columns.

```python
# Category with Higher Unit Sales Last Year Compared to Current Year
# Compare unit sales last year and current year for each category
unit_sales_comparison = data_alcohol.groupby('Category').agg({'Unit sales': 'sum', 'Unit sales last year': 'sum'})

# Filter categories where unit sales last year are higher than the current year
higher_unit_sales_last_year = unit_sales_comparison[unit_sales_comparison['Unit sales last year'] > unit_sales_comparison['Unit sales']]

# Display the result
if not higher_unit_sales_last_year.empty:
    print("Categories with higher unit sales last year compared to the current year:")
    print(higher_unit_sales_last_year)
else:
    print("No category had higher unit sales last year compared to the current year.")
```

# Category with the Highest Number of Units Sold Among Selected Categories

In this section, we will determine which category among Beverages, Grains, Fruits, and Alcohol has the highest number of units sold (`Unit sales`).

```python
# Category with the Highest Number of Units Sold Among Selected Categories
# Filter the dataset for the selected categories
selected_categories = ['Beverages', 'Grains', 'Fruits', 'Alcohol']
selected_data = data_alcohol[data_alcohol['Category'].isin(selected_categories)]

# Group by category and calculate total unit sales
category_unit_sales = selected_data.groupby('Category')['Unit sales'].sum()

# Find the category with the highest unit sales
highest_unit_sales_category = category_unit_sales.idxmax()
highest_unit_sales = category_unit_sales.max()

# Display the result
print(f'The category with the highest number of units sold is \"{highest_unit_sales_category}\" with {highest_unit_sales} units sold.')
```
